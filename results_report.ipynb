```python
import pickle
import pandas as pd
import seaborn; seaborn.set()
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import display
import glob
from IPython.core.display import HTML

pd.set_option('display.max_colwidth', -1)
```

```python
exp_name = "may14"
results_dir = "experiment_output/master_summary_dataframe-20180514-190154-clipped/"
#results_dir = "experiment_output/master_summary_dataframe-20180514-190154/"

files = glob.glob(results_dir + exp_name+"*.pkl")

df_results_list_daily    = []
df_results_list_intake    = []

df_results      = []
for pkl_file in files:
    try:
        if(exp_name in pkl_file):
            results = pickle.load( open(pkl_file, "rb" ) )
            if(".d." in pkl_file):
                df_results_list_daily.append(results["df_perf"])
            else:
                df_results_list_intake.append(results["df_perf"])                
    except:
        continue


if(len(df_results_list_intake)>0):
    df_results_intake = pd.concat(df_results_list_intake)
    vals = df_results_intake["Test R"].values
    ind = np.argsort(-1*np.array(vals))
    df_results_intake =df_results_intake.sort_values("Test R",ascending=False)

    display(HTML("<H2>Intake Results Summary Report</H2>"))
    display(df_results_intake)
    display(HTML("<BR><HR style='border-top: 1px solid black;'>"))
    

if(len(df_results_list_daily)>0):
    df_results_daily = pd.concat(df_results_list_daily)
    vals = df_results_daily["Test R"].values
    ind = np.argsort(-1*np.array(vals))
    df_results_daily=df_results_daily.sort_values("Test R",ascending=False)

    display(HTML("<H2>Daily Results Summary Report</H2>"))
    display(df_results_daily)
    display(HTML("<BR><HR style='border-top: 1px solid black;'>"))




for pkl_file in np.array(files):
    #try:
        results = pickle.load( open(pkl_file, "rb" ) )
        indciator_name = results["df_config"]["Value"][0]

        display(HTML("<H2>Results Detail Report for %s</H2>"%(indciator_name) ))

        display(HTML("<H3>Prediction Performance</H3>"))    
        display(results["df_perf"])

        plt.figure(figsize=(10,8))
        plt.subplot(2,2,1)
        results["df_tr"]["target"].plot.hist() 
        plt.title("Train True Hist")

        plt.subplot(2,2,2)
        results["df_te"]["target"].plot.hist() 
        plt.title("Test True Hist")

        plt.subplot(2,2,3)
        results["df_tr"]["prediction"].plot.hist() 
        plt.title("Train Prediction Hist")

        plt.subplot(2,2,4)
        results["df_te"]["prediction"].plot.hist() 
        plt.title("Test Prediction Hist")    

        plt.figure(figsize=(10,4))    
        plt.subplot(1,2,1)
        plt.plot(np.array(results["df_tr"]["prediction"]),np.array(results["df_tr"]["target"]),"bo",alpha=0.2)
        plt.xlabel("Prdicted")
        plt.ylabel("True")
        plt.title("Train True vs Prediction")

        plt.subplot(1,2,2)
        plt.plot(np.array(results["df_te"]["prediction"]),np.array(results["df_te"]["target"]),"bo",alpha=0.2)
        plt.xlabel("Prdicted")
        plt.ylabel("True")
        plt.title("Test True vs Prediction")   
        plt.show()

        display(HTML("<H3>Selected Features</H3>"))    
        display(results["df_features"])

        selected = np.array(results["df_features"]["Features"])
        y_tr = np.array(results["df_tr"]["target"])
        y_te = np.array(results["df_te"]["target"]) 
        #y_raw = np.array(results["df_raw"]["target"])

        if(0): 
            for f in selected:

                display(HTML("<B>Featrue-Target Relationship for %s</B>"%(f)))
                if(f in list(results["df_raw"].columns)):
                    x_raw = np.array(results["df_raw"][f])
                else:
                    x_raw  = y_raw*np.nan
                x_tr = np.array(results["df_tr"][f])
                x_te = np.array(results["df_te"][f])
                xmin,xmax = min(x_tr)-1,max(x_tr)+1    

                plt.figure(figsize=(15,4)) 
                plt.subplot(1,3,1)    
                plt.plot(x_raw,y_raw ,"bo",alpha=0.2)
                plt.xlabel("Feature")
                plt.ylabel("Target")
                plt.xlim(xmin,xmax)
                plt.title("Raw Data")

                plt.subplot(1,3,2)
                plt.plot(x_tr,y_tr,"bo",alpha=0.2)
                plt.xlabel("Feature")
                plt.ylabel("Target")
                plt.xlim(xmin,xmax)
                plt.title("Training Data")

                plt.subplot(1,3,3)
                plt.plot(x_te,y_te,"bo",alpha=0.2) 
                plt.xlabel("Feature")
                plt.ylabel("Target")
                plt.xlim(xmin,xmax)
                plt.title("Test Data")
                plt.show()    


        display(HTML("<H3>Experiment Configuration</H3>"))    
        display(results["df_config"])

        display(HTML("<BR><HR style='border-top: 1px solid black;'><BR><BR>"))
    #except:
    #    continue
```
