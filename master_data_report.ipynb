```python
#reload(master_data_report)
pd.options.display.max_rows=10000
pd.options.display.max_colwidth=-1

dump_dir = "master_data_file"
dump_file="master_summary_dataframe-20180514-143842.pkl"    
out = pickle.load( open(dump_dir + "/" + dump_file, "rb" ) )
df=out["dataframe"]
md=out["metadata"]
#df=out

sumdf,missudf = master_data_report.summary_report(df)

display(HTML("<H2>Master Summary Data Report</H2>"))

display(HTML("<H3>Data Summary</H3>"))
display(sumdf)

display(HTML("<H3>Users With No Data</H3>"))
display(missudf)

df_user = master_data_report.subject_report(df)
display(HTML("<H3>User Report</H3>"))
display(df_user)

display(HTML("<H3>Indicator Report</H3>"))
df_feature = master_data_report.feature_report(df)
df_feature.index.rename("Indicator",inplace=True)
ind = [ "data_qualtrics" in x for x in list(df_feature.index)]
display(df_feature[ind])

display(HTML("<H3>Marker Report</H3>"))
df_feature.index.rename("Marker",inplace=True)
ind = [ "data_qualtrics" not in x for x in list(df_feature.index)]
display(df_feature[ind])

display(HTML("<H3>Meta Data Report</H3>"))
streams = df.columns
mdf=pd.DataFrame()
for stream in md.keys():
    if("md2k" in stream): 
        #stream = stream.split("(")[0]
        data = [[stream]+list(md[stream].values())]
        cols = ["Stream"]+list(md[stream].keys())
        mdf=mdf.append(pd.DataFrame(data=data,columns=cols))
display(mdf)   

```
